### Table 3.7 Describing the home team – away team goal difference (and 3.9)

```python
pd.DataFrame.from_dict(
    {
        "Statistics": [
            "Mean",
            "Standard deviation",
            "Percent positive",
            "Percent zero",
            "Percent negative",
            "Number of observations",
        ],
        "Value": [
            df["home_goaladv"].describe()["mean"],
            df["home_goaladv"].describe()["std"],
            (df["home_goaladv"] > 0).sum() / df["home_goaladv"].shape[0] * 100,
            (df["home_goaladv"] == 0).sum() / df["home_goaladv"].shape[0] * 100,
            (df["home_goaladv"] < 0).sum() / df["home_goaladv"].shape[0] * 100,
            df["home_goaladv"].describe()["count"],
        ],
    }
).round(1)
```

```python
df["home_goaladv"].value_counts()
```

```python
perc = df["home_goaladv"].value_counts(normalize=True).sort_index() * 100
goal_diff = sorted(df["home_goaladv"].dropna().unique())

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(goal_diff, perc.values, color="navy", edgecolor="white", width=0.9)


# Add labels above bars
ax.bar_label(bars, labels=[f"{val:.1f}%" for val in perc.values], padding=3)

# Formatting
ax.set_xlabel("Goals")
ax.set_ylabel("Share of Ganes (percent %)")
ax.set_title("Goal Difference (Home - Away)")
ax.set_xticks(goal_diff)
ax.set_ylim(0, perc.max() + 10)
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```

### Figure 3.9 The distribution of home team – away team goal difference

```python
from plotnine import ggplot, aes, geom_histogram, geom_text, labs, scale_x_continuous, scale_y_continuous, theme_bw
from mizani.formatters import percent_format
import numpy as np

# The code in cell 17 is incomplete and contains errors.
# Here is a corrected version using plotnine to plot the distribution of home_goaladv as a histogram with percentage labels.


# Prepare data for labels
hist_data = df['home_goaladv'].value_counts(normalize=True).sort_index().reset_index()
hist_data.columns = ['home_goaladv', 'proportion']
hist_data['label'] = hist_data['proportion'].apply(lambda x: f"{x*100:.1f}%")
hist_data['proportion'] = hist_data['proportion']  # already in proportion

(
    ggplot(df, aes(x="home_goaladv"))
    + geom_histogram(aes(y="..count../sum(..count..)"), binwidth=1, fill="navy", color="white")
    + geom_text(
        data=hist_data,
        mapping=aes(x="home_goaladv", y="proportion", label="label"),
        va='bottom',
        size=10,
        nudge_y=0.01
    )
    + labs(x="Goal difference", y="Share of games (percent)")
    + scale_x_continuous(expand=(0.05, 0.05), limits=(-6, 6), breaks=np.arange(-6, 7, 1))
    + scale_y_continuous(
        expand=(0, 0),
        limits=(0, 0.25),
        breaks=np.arange(0, 0.26, 0.05),
        labels=percent_format()
    )
    + theme_bw()
)
```
