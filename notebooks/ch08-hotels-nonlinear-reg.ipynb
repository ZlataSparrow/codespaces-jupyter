```python
import os
import sys
import warnings

import numpy as np
import pandas as pd
from mizani.formatters import percent_format
from plotnine import *
import statsmodels.formula.api as smf

warnings.filterwarnings("ignore")
```

```python
# load vienna
hotels = pd.read_csv("/workspaces/codespaces-jupyter/data/hotels-vienna.csv")
#data = pd.read_csv("https://osf.io/y6jvb/download")
```

```python
hotels = hotels.loc[
    (hotels["accommodation_type"] == "Hotel")
    & (hotels["city_actual"] == "Vienna")
    & (hotels["stars"] >= 3)
    & (hotels["stars"] <= 4)
    & (hotels["stars"].notna())
    & (hotels["price"] <= 600)
]
```

#### LOG MODELS

```python
hotels["lnprice"] = np.log(hotels["price"])
hotels["distance2"] = hotels["distance"]
hotels.loc[hotels["distance2"] < 0.05, "distance2"] = 0.05
hotels["lndistance"] = np.log(hotels["distance2"])
```

```python
hotels["price"].describe()
```

```python
hotels["lnprice"].describe()
```

## Regression

```python
# LEVEL - LEVEL
reg1 = smf.ols("price ~ distance", data=hotels).fit()
reg1.summary()
```

```python
# LEVEL - LOG
reg2 = smf.ols("price ~ lndistance", data=hotels).fit()
reg2.summary()
```

```python
# LOG - LEVEL
reg3 = smf.ols("lnprice ~ distance", data=hotels).fit()
reg3.summary()
```

```python
# LOG - LOG
reg4 = smf.ols("lnprice ~ lndistance", data=hotels).fit()
reg4.summary()
```

### Figure 8.1 Regressions of hotel price and distance to the city center

```python
(
    ggplot(hotels, aes(x="distance", y="price"))
    + geom_point(color="blue")
    + geom_smooth(method="lm", color="green", se=False)
    + expand_limits(x=0.01, y=0.01)
    + scale_x_continuous(expand=[0.01, 0.01], limits=[0, 7], breaks=np.arange(0, 8, 1))
    + scale_y_continuous(expand=[0.01, 0.01], limits=[0, 400], breaks=np.arange(0, 400, 50))
    + labs(x="Distance to city center (miles) (LEVEL - LEVEL)", y="Price (US dollars)")
    + theme_bw()
)
```

```python
(
    ggplot(hotels, aes(x="distance", y="lnprice"))
    + geom_point(color="blue")
    + geom_smooth(method="lm", color="green", se=False)
    + expand_limits(x=0.01, y=0.01)
    + scale_x_continuous(expand=[0.01, 0.01], limits=[0, 7], breaks=np.arange(0, 8, 1))
    + scale_y_continuous(
        expand=[0.01, 0.01], limits=[3.5, 6], breaks=np.arange(3.5, 6, 0.5)
    )
    + labs(x="Distance to city center (miles) (LOG - LEVEL", y="ln(price, US dollars)")
    + theme_bw()
)
```

```python
stargazer = Stargazer([reg1, reg3, reg2, reg4])
stargazer.covariate_order(["distance", "lndistance", "Intercept"])
stargazer.custom_columns(["price", "ln(price)", "price", "ln(price)"], [1, 1, 1, 1])
stargazer.rename_covariates(
    {
        "Intercept": "Constant",
        "distance": "Distance to city center, miles",
        "lndistance": "ln(distance to city center)",
    }
)
stargazer
```
