```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
import pandas as pd
import numpy as np
np.random.seed(42)
df = pd.DataFrame({
    "x": np.random.normal(0, 1, 1000),
    "y": np.random.normal(0, 1, 1000) + 0.5 * np.random.normal(0, 1, 1000)
})

# Jointplot
sns.jointplot(data=df, x="x", y="y", kind="scatter", alpha=0.5)
plt.show()
```

```python
sns.jointplot(data=df, x="x", y="y", kind="kde", fill=True, cmap="Blues")
plt.show()
```
